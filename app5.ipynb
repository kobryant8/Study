```python
시작 = 1
끝 = 100
while 시작 < 끝:
    if 시작%2 == 0:
        print(f"출력 : {시작}")

    선택한음료 = input('버튼을 눌러 음료수를 선택해 주세요: ')

    if 선택한음료 == '콜라' or 선택한음료 == '칠성사이다' or 선택한음료 == '환타' or 선택한음료 == '레몬녹차':

        if 선택한음료 == '콜라':
            금액 = 1700
            print('음료 선택 콜라 1700원')

        elif 선택한음료 == '칠성사이다':
            금액 = 1500
            print('음료 선택 칠성사이다 1500원')

        elif 선택한음료 == '환타':
            금액 = 1600
            print('음료 선택 환타 1600원')

        elif 선택한음료 == '레몬녹차':
            금액 = 1000
            print('음료 선택 레몬녹차 1000원')

        수량 = int(input('수량을 입력하세요: '))

        총금액 = 금액 * 수량
        print(f'{선택한음료} {수량}개 금액은 {총금액}원 입니다.')

        지불금액 = int(input('지불할 금액을 입력하세요: '))
        잔돈 = 지불금액 - 총금액
        print(f'{지불금액}원을 받고 {잔돈}원을 잔돈으로 뽑아주세요.')

    else:
        print('다른 음료는 판매하지 않습니다.')
```

```python
while 1:
    print("음료수를 구매할 수 있는 자판기입니다")
    음료 = input("콜라, 사이다, 환타 중 하나를 주문하세요.(종료는 'exit'를 입력하세요.)")
    if 음료 == '콜라' or 음료 == '사이다' or 음료 == '환타':
    if 음료 == 'exit':
        print("자판기가 종료됩니다.")
        break
    금액 = 0
    if 음료 == '콜라':
        print('콜라를 주문했습니다. 500원')
        금액 = 500
    elif 음료 == '사이다':
        print('사이다를 주문했습니다. 450원')
        금액 = 450
    elif 음료 == '환타':
        print('환타를 주문했습니다. 300원')
        금액 = 300
        
        수량 = input('수량을 입력하세요.')
        
        총금액 = 금액 * int(수량)
        print(f'{음료} {수량}개 금액은 {총금액}원입니다.')
        
        지불금액 = input('지불 금액을 입력하세요.')
        거스름돈 = int(지불금액) - (총금액)
        print(f'{지불금액}원을 받고 {거스름돈}원 거스름돈을 받으세요.')
```
